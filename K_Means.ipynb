# K_MEANS

```python
def initiate_centroids(k, X): # k: no. of clusters, X: dataset multidimensional array
    arr = []              # array of initial multidimensional centroids
    for i in range(k):
        cxj = []          # array of multiple dimensions point
        for j in range(X.shape[1]):
            cx = np.random.uniform(min(X[:,j]), max(X[:,j])) # single dimension picked randomly from a uniform distribution within min to max range
            cxj.append(cx)
        arr.append(cxj)
    return np.asarray(arr)

#init_centroids(3,X)
```

```python
def euclidian_dist(a, b):           # euclidian distance between two points
    return np.sqrt(sum(np.square(a-b)))

#euclidian_dist(X[795,:],X[254,:])
```

```python
def assign_cluster(k, X, centroids): # cg: array of centroids
    cluster = [-1]*len(X)     # cluster array representing every entry's cluster, initialised with -1
    for i in range(len(X)):   # for every entry in dataset
        dist_arr = []
        for j in range(k):    # for every cluster/centroid 
            dist_arr.append(euclidian_dist(X[i], centroids[j]))   # add dist of each point from every centroid
        idx = np.argmin(dist_arr)  # getting the indexes of that centroid whose distance is minimum from particular point
        cluster[i] = idx           # storing index, ith entry has been alloted to prev calculated min dist centroid point
    return np.asarray(cluster)

#assign_cluster(3,X,init_centroids(3,X))
```

```python
def compute_centroids(k, X, cluster):
    centroids_arr = []              # initialising new array of centroids
    for i in range(k):
        arr = []             # array of points belonging to particular cluster
        for j in range(len(X)):
            if cluster[j]==i:
                arr.append(X[j])
        centroids_arr.append(np.mean(arr, axis=0))     #finding new centroid by taking mean of all points in that particular clusters
    return np.asarray(centroids_arr)

#compute_centroids(3,X,assign_cluster(3,X,init_centroids(3,X)))
```

```python
def measure_change(centroids_prev, centroids_new):     # arrays containing new and old centroid points
    res = 0
    for a,b in zip(centroids_prev,centroids_new):      # zip helps to iterate over two arrays simultaneously
        res+=euclidian_dist(a,b)         # computing and summing distance between new and old centroids respectively
    return res
```

```python
## driving function for the k-means algorithm as a whole on a dataset given as an array
def k_means(k, X):
    centroids_prev = initiate_centroids(k, X)
    cluster = [0]*len(X)
    centroids_change = 100
    i=0
    while centroids_change>.000001:
        cluster = assign_cluster(k, X, centroids_prev)
        #show_clusters(X, cluster, centroids_prev)
        centroids_new = compute_centroids(k, X, cluster)
        centroids_change = measure_change(centroids_new, centroids_prev)
        #print(centroids_change)
        centroids_prev = centroids_new
        i+=1
    print("No. of iterations applied: ",i,"\n")
    return cluster,centroids_prev
```

```python
## Taking k = 4, as after running elbow method on dataset, it is found to be elbow point 
## hence further increase in k value will result in less intercluster distance and more intracluster distance
## Above point is verified by keeping k-value higher than the 4 and resulting clusters when visualised on 2-D plane were intertwined into one another
k = 4
clusters,centroids = k_means(k, X)

#print(centroids)
#printing clusters

cluster_n = set(clusters)

clusters_arr = []
for value in cluster_n:
    cluster_i= []
    for index in range(len(clusters)):
        if clusters[index] == value:
            cluster_i.append(index)
    clusters_arr.append(cluster_i)

for value in cluster_n:
    print("Cluster",value,": ",clusters_arr[value],"\n")
```

```python
## Using PCA dimensionality reduction only to view clusters in a 2-D plane

from sklearn.decomposition import PCA
df = pd.DataFrame(X)
centroids = pd.DataFrame(centroids)
cols = df.columns[:-1]
df['Cluster'] = clusters

pca = PCA(n_components=2)

df['x'] = pca.fit_transform(df[cols])[:,0]
df['y'] = pca.fit_transform(df[cols])[:,1]
centroids['x'] = pca.fit_transform(centroids[cols])[:,0]
centroids['y'] = pca.fit_transform(centroids[cols])[:,1]

fig = plt.figure(figsize=(15, 15))
plt.scatter(df['x'], df['y'], c=df['Cluster'], marker='o', s=20)
plt.scatter(centroids['x'],centroids['y'], c='black', marker="+",s=100)
```
